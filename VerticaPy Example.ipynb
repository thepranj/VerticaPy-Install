The [Telco Customer Churn](https://github.com/thepranj/VerticaPy-Install/blob/main/churn.csv) dataset (also available on [Kaggle](https://www.kaggle.com/blastchar/telco-customer-churn)) contains information on Telco customers and their contracts. Our goal is to predict what factors lead customers to "churn," that is, leave Telco.

To load churn.csv into the database:

```python
# Drop the table public.churn in case it already exists
drop_table(name = 'public.churn')
# Create public.churn with the data in churn.csv
read_csv("churn.csv",
        schema = 'public',
        table_name = 'churn')
        #genSQL = True
```
